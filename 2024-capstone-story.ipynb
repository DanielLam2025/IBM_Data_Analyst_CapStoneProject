```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the survey dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Optional: remove leading/trailing spaces in column names
df.rename(columns=lambda x: x.strip(), inplace=True)

# Check first few rows
print(df.head())

print(df.columns.tolist())

# Drop missing values
df_db = df.dropna(subset=["DatabaseHaveWorkedWith"]).copy()

# Split multiple databases into separate rows
df_db = df_db.assign(Database=df_db["DatabaseHaveWorkedWith"].str.split(";"))
df_db = df_db.explode("Database")
df_db["Database"] = df_db["Database"].str.strip()

# Count top 10 databases
top_databases = df_db["Database"].value_counts().head(10)

# Plot
plt.figure(figsize=(8,6))
top_databases.plot(kind='bar', color='salmon', edgecolor='black')
plt.title("Top 10 Databases Respondents Have Worked With")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Print all column names to see which database columns are available
for col in df.columns:
    print(col)
df_db_future = df.dropna(subset=["DatabaseWantToWorkWith"]).copy()  # if exists
df_db_future = df_db_future.assign(Database=df_db_future["DatabaseWantToWorkWith"].str.split(";"))
df_db_future = df_db_future.explode("Database")
df_db_future["Database"] = df_db_future["Database"].str.strip()
top_databases_future = df_db_future["Database"].value_counts().head(10)
top_databases_future.plot(kind='bar', color='lightblue', edgecolor='black')
plt.title("Top 10 Databases Respondents Want to Work With Next Year")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Map age groups to numeric midpoints
age_mapping = {
    "Under 18 years old": 16,
    "18-24 years old": 21,
    "25-34 years old": 29.5,
    "35-44 years old": 39.5,
    "45-54 years old": 49.5,
    "55-64 years old": 59.5,
    "65 years or older": 70
}
df['AgeNumeric'] = df['Age'].map(age_mapping)

# Median compensation by age
median_comp = df.dropna(subset=["ConvertedCompYearly","AgeNumeric"]).groupby("AgeNumeric")["ConvertedCompYearly"].median()

plt.figure(figsize=(8,6))
median_comp.plot(kind='bar', color='purple', edgecolor='black')
plt.title("Median Compensation by Age Group")
plt.xlabel("Age")
plt.ylabel("Median Yearly Compensation (USD)")
plt.tight_layout()
plt.show()
```

```python
# Drop missing values
df_exp = df.dropna(subset=["WorkExp","JobSatPoints_6"]).copy()

# Median job satisfaction by experience
median_jobsat = df_exp.groupby("WorkExp")["JobSatPoints_6"].median()

plt.figure(figsize=(10,6))
median_jobsat.plot(kind='line', marker='o', color='green')
plt.title("Median Job Satisfaction vs Experience")
plt.xlabel("Years of Professional Experience")
plt.ylabel("Median Job Satisfaction (JobSatPoints_6)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df_bubble = df.dropna(subset=["ConvertedCompYearly","JobSatPoints_6","AgeNumeric"]).copy()

plt.figure(figsize=(10,6))
scatter = plt.scatter(
    df_bubble['ConvertedCompYearly'],
    df_bubble['JobSatPoints_6'],
    s=df_bubble['AgeNumeric']*2,   # bubble size
    c=df_bubble['AgeNumeric'],     # color = age
    cmap='viridis',
    alpha=0.5,
    edgecolor='k'
)
plt.title("Compensation vs Job Satisfaction (Bubble = Age)")
plt.xlabel("Yearly Compensation (USD)")
plt.ylabel("Job Satisfaction (JobSatPoints_6)")
plt.colorbar(scatter, label="Age")
plt.xlim(0, df_bubble['ConvertedCompYearly'].quantile(0.95))  # focus on 95% data
plt.tight_layout()
plt.show()
```
