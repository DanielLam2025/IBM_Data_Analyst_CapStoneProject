### Task 1: Stacked Chart for Composition of Job Satisfaction Across Age Groups


##### 1. Stacked Chart of Median `JobSatPoints_6` and `JobSatPoints_7` for Different Age Groups


Visualize the composition of job satisfaction scores (`JobSatPoints_6` and `JobSatPoints_7`) across various age groups. This will help in understanding the breakdown of satisfaction levels across different demographics.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("survey-data.csv")

# Preview the data
print(df.head())

# Group data by Age group and calculate median job satisfaction points
age_group_job_sat = df.groupby("Age")[
    ["JobSatPoints_6", "JobSatPoints_7"]
].median()

# Plot stacked bar chart
ax = age_group_job_sat.plot(
    kind="bar",
    stacked=True,
    figsize=(10, 6)
)

# Formatting
plt.title("Median Job Satisfaction Points (6 & 7) Across Age Groups", fontsize=14)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Median Job Satisfaction Points", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Job Satisfaction Points")
plt.tight_layout()

plt.show()
```

##### Stacked Chart of `JobSatPoints_6` and `JobSatPoints_7` for Employment Status


Create a stacked chart to compare job satisfaction (`JobSatPoints_6` and `JobSatPoints_7`) across different employment statuses. This will show how satisfaction varies by employment type.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("survey-data.csv")

# Group by Employment Status and calculate median job satisfaction points
employment_job_sat = df.groupby("Employment")[
    ["JobSatPoints_6", "JobSatPoints_7"]
].median()

# Plot stacked bar chart
ax = employment_job_sat.plot(
    kind="bar",
    stacked=True,
    figsize=(10, 6),
)

# Formatting
plt.title("Median Job Satisfaction Points (6 & 7) Across Employment Status", fontsize=14)
plt.xlabel("Employment Status", fontsize=12)
plt.ylabel("Median Job Satisfaction Points", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Job Satisfaction Points")
plt.tight_layout()

plt.show()
```

### Task 2: Stacked Chart for Compensation and Job Satisfaction by Age Group

### Task 3: Comparing Data Using Stacked Charts


##### 1. Stacked Chart of Preferred Databases by Age Group




Visualize the top databases that respondents from different age groups wish to learn. Create a stacked chart to show the proportion of each database in each age group.


```python
# Split multi-database responses into separate rows
df_db = df.dropna(subset=["DatabaseWantToWorkWith"]).copy()
df_db = df_db.assign(Database=df_db["DatabaseWantToWorkWith"].str.split(";"))
df_db = df_db.explode("Database")

# Clean whitespace
df_db["Database"] = df_db["Database"].str.strip()

# Count database preferences per age group
db_counts = df_db.groupby(["Age", "Database"]).size().unstack(fill_value=0)

# Normalize to proportions
db_props = db_counts.div(db_counts.sum(axis=1), axis=0)

# Plot stacked bar chart
ax = db_props.plot(
    kind="bar",
    stacked=True,
    figsize=(12, 7)
)

plt.title("Preferred Databases by Age Group (Proportions)", fontsize=14)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Proportion of Respondents", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Database", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

##### 2. Stacked Chart of Employment Type by Job Satisfaction


Analyze the distribution of employment types within each job satisfaction level using a stacked chart. This will provide insights into how employment types are distributed across various satisfaction ratings.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("survey-data.csv")

# Drop rows with missing values in Employment or JobSatPoints_6
df_filtered = df.dropna(subset=["Employment", "JobSatPoints_6"]).copy()

# Count employment types per job satisfaction score
employment_counts = df_filtered.groupby(["JobSatPoints_6", "Employment"]).size().unstack(fill_value=0)

# Normalize to proportions (optional, for easier comparison)
employment_props = employment_counts.div(employment_counts.sum(axis=1), axis=0)

# Plot stacked bar chart
ax = employment_props.plot(
    kind="bar",
    stacked=True,
    figsize=(12, 7),
)

# Formatting
plt.title("Employment Type Distribution by Job Satisfaction", fontsize=14)
plt.xlabel("Job Satisfaction Score (JobSatPoints_6)", fontsize=12)
plt.ylabel("Proportion of Respondents", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Employment Type", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()

plt.show()
```

### Task 4: Exploring Technology Preferences Using Stacked Charts


##### 1. Stacked Chart for Preferred Programming Languages by Age Group


Analyze how programming language preferences (`LanguageAdmired`) vary across age groups.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("survey-data.csv")

# Check column names to find the correct language column
print(df.columns.tolist())  # look for something like 'LanguageAdmired'

# Drop rows with missing language preferences
df_lang = df.dropna(subset=["LanguageAdmired"]).copy()

# Split multiple languages into separate rows
df_lang = df_lang.assign(Language=df_lang["LanguageAdmired"].str.split(";"))
df_lang = df_lang.explode("Language")

# Clean whitespace
df_lang["Language"] = df_lang["Language"].str.strip()

# Count language preferences per age group
lang_counts = df_lang.groupby(["Age", "Language"]).size().unstack(fill_value=0)

# Normalize to proportions for better comparison
lang_props = lang_counts.div(lang_counts.sum(axis=1), axis=0)

# Plot stacked bar chart
ax = lang_props.plot(
    kind="bar",
    stacked=True,
    figsize=(12, 7)
)

# Formatting
plt.title("Preferred Programming Languages by Age Group (Proportions)", fontsize=14)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Proportion of Respondents", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Language", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()

plt.show()
```

##### 2. Stacked Chart for Technology Adoption by Employment Type


Explore how admired platforms (`PlatformAdmired`) differ across employment types (e.g., full-time, freelance)


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("survey-data.csv")

# Check column names to find the correct platform column
print(df.columns.tolist())  # look for something like 'PlatformAdmired'

# Drop rows with missing values in PlatformAdmired or Employment
df_platform = df.dropna(subset=["PlatformAdmired", "Employment"]).copy()

# Split multiple platforms into separate rows
df_platform = df_platform.assign(Platform=df_platform["PlatformAdmired"].str.split(";"))
df_platform = df_platform.explode("Platform")

# Clean whitespace
df_platform["Platform"] = df_platform["Platform"].str.strip()

# Count platform preferences per employment type
platform_counts = df_platform.groupby(["Employment", "Platform"]).size().unstack(fill_value=0)

# Normalize to proportions for comparison
platform_props = platform_counts.div(platform_counts.sum(axis=1), axis=0)

# Plot stacked bar chart
ax = platform_props.plot(
    kind="bar",
    stacked=True,
    figsize=(12, 7)
)

# Formatting
plt.title("Technology/Platform Adoption by Employment Type (Proportions)", fontsize=14)
plt.xlabel("Employment Type", fontsize=12)
plt.ylabel("Proportion of Respondents", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Platform", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()

plt.show()
```
