### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("survey-data.csv")

# Drop rows with missing compensation or age
df_filtered = df.dropna(subset=["ConvertedCompYearly", "Age"]).copy()

# Group by Age and calculate median compensation
median_comp = df_filtered.groupby("Age")["ConvertedCompYearly"].median()

# Sort age groups in logical order
age_order = ["Under 18 years old", "18-24 years old", "25-34 years old",
             "35-44 years old", "45-54 years old", "55-64 years old", "65 years or older"]
median_comp = median_comp.reindex(age_order)

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(median_comp.index, median_comp.values, marker='o', linestyle='-', color='blue')
plt.title("Median Yearly Compensation by Age Group", fontsize=14)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Median Compensation (USD)", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("survey-data.csv")

# Filter for relevant age groups
age_groups_25_45 = ["25-34 years old", "35-44 years old"]
df_filtered = df[df["Age"].isin(age_groups_25_45)].dropna(subset=["ConvertedCompYearly"])

# Group by Age and calculate median compensation
median_comp = df_filtered.groupby("Age")["ConvertedCompYearly"].median()

# Ensure logical order
median_comp = median_comp.reindex(age_groups_25_45)

# Plot line chart
plt.figure(figsize=(8, 5))
plt.plot(median_comp.index, median_comp.values, marker='o', linestyle='-', color='green')
plt.title("Median Yearly Compensation for Ages 25-45", fontsize=14)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Median Compensation (USD)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("survey-data.csv")

# Use YearsCodePro as proxy for experience
df_exp = df.dropna(subset=["YearsCodePro", "JobSatPoints_6"]).copy()

# Convert YearsCodePro to numeric
def convert_years(x):
    if x == "Less than 1 year":
        return 0.5
    elif x == "More than 50 years":
        return 51
    else:
        try:
            return float(x)
        except:
            return None

df_exp["YearsCodeProNumeric"] = df_exp["YearsCodePro"].apply(convert_years)
df_exp = df_exp.dropna(subset=["YearsCodeProNumeric"])

# Group by experience and calculate median job satisfaction
median_jobsat = df_exp.groupby("YearsCodeProNumeric")["JobSatPoints_6"].median()

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(median_jobsat.index, median_jobsat.values, marker='o', linestyle='-', color='purple')
plt.title("Median Job Satisfaction (JobSatPoints_6) by Experience Level", fontsize=14)
plt.xlabel("Years of Professional Experience", fontsize=12)
plt.ylabel("Median Job Satisfaction Points", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience
